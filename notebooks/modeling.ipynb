```python
"""
Task 2 - Model Building and Training
Improved Detection of Fraud Cases for E-commerce and Bank Transactions
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Machine Learning imports
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    average_precision_score,
    precision_recall_curve,
    f1_score,
    precision_score,
    recall_score,
    roc_curve
)
from sklearn.metrics import ConfusionMatrixDisplay
import joblib

print("="*60)
print("TASK 2: MODEL BUILDING AND TRAINING")
print("="*60)

# ============================================================================
# 1. LOAD PREPARED DATA
# ============================================================================

print("\n" + "="*60)
print("LOADING PREPARED DATA")
print("="*60)

# Load fraud data
fraud_train = joblib.load('../data/processed/fraud_train_resampled.pkl')
fraud_test = joblib.load('../data/processed/fraud_test.pkl')

# Load credit card data
credit_train = joblib.load('../data/processed/credit_train_resampled.pkl')
credit_test = joblib.load('../data/processed/credit_test.pkl')

# Unpack
X_fraud_train, y_fraud_train = fraud_train
X_fraud_test, y_fraud_test = fraud_test
X_credit_train, y_credit_train = credit_train
X_credit_test, y_credit_test = credit_test

# Load feature info
feature_info = joblib.load('../data/processed/feature_info.pkl')
fraud_features = feature_info['fraud_features']
credit_features = feature_info['credit_features']

print(f"\nFraud Data:")
print(f"  Training: {X_fraud_train.shape}, Fraud rate: {y_fraud_train.mean()*100:.2f}%")
print(f"  Test:     {X_fraud_test.shape}, Fraud rate: {y_fraud_test.mean()*100:.2f}%")

print(f"\nCredit Card Data:")
print(f"  Training: {X_credit_train.shape}, Fraud rate: {y_credit_train.mean()*100:.2f}%")
print(f"  Test:     {X_credit_test.shape}, Fraud rate: {y_credit_test.mean()*100:.2f}%")

# ============================================================================
# 2. EVALUATION FUNCTIONS
# ============================================================================

print("\n" + "="*60)
print("DEFINING EVALUATION FUNCTIONS")
print("="*60)

def evaluate_model(model, X_train, y_train, X_test, y_test, model_name, dataset_name):
    """
    Comprehensive model evaluation function
    """
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    # Calculate metrics
    metrics = {
        'Model': model_name,
        'Dataset': dataset_name,
        'Accuracy': (y_pred == y_test).mean(),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'AUC-ROC': roc_auc_score(y_test, y_pred_proba),
        'AUC-PR': average_precision_score(y_test, y_pred_proba)
    }
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    metrics['True Negatives'] = tn
    metrics['False Positives'] = fp
    metrics['False Negatives'] = fn
    metrics['True Positives'] = tp
    
    return metrics, y_pred, y_pred_proba

def plot_confusion_matrix(y_test, y_pred, model_name, dataset_name):
    """Plot confusion matrix"""
    fig, ax = plt.subplots(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
    ax.set_title(f'Confusion Matrix - {model_name} ({dataset_name})')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
    plt.tight_layout()
    return fig

def plot_roc_pr_curves(y_test, y_pred_proba, model_name, dataset_name):
    """Plot ROC and Precision-Recall curves"""
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    axes[0].plot(fpr, tpr, label=f'ROC (AUC = {roc_auc:.3f})')
    axes[0].plot([0, 1], [0, 1], 'k--')
    axes[0].set_xlabel('False Positive Rate')
    axes[0].set_ylabel('True Positive Rate')
    axes[0].set_title(f'ROC Curve - {model_name} ({dataset_name})')
    axes[0].legend()
    axes[0].grid(True)
    
    # Precision-Recall Curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    pr_auc = average_precision_score(y_test, y_pred_proba)
    axes[1].plot(recall, precision, label=f'PR (AUC = {pr_auc:.3f})')
    axes[1].set_xlabel('Recall')
    axes[1].set_ylabel('Precision')
    axes[1].set_title(f'Precision-Recall Curve - {model_name} ({dataset_name})')
    axes[1].legend()
    axes[1].grid(True)
    
    plt.tight_layout()
    return fig

def cross_validate_model(model, X, y, model_name, dataset_name, cv=5):
    """Perform cross-validation and return scores"""
    cv = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
    
    scores = {
        'Model': model_name,
        'Dataset': dataset_name,
        'AUC-ROC_mean': cross_val_score(model, X, y, cv=cv, scoring='roc_auc').mean(),
        'AUC-ROC_std': cross_val_score(model, X, y, cv=cv, scoring='roc_auc').std(),
        'AUC-PR_mean': cross_val_score(model, X, y, cv=cv, scoring='average_precision').mean(),
        'AUC-PR_std': cross_val_score(model, X, y, cv=cv, scoring='average_precision').std(),
        'F1_mean': cross_val_score(model, X, y, cv=cv, scoring='f1').mean(),
        'F1_std': cross_val_score(model, X, y, cv=cv, scoring='f1').std()
    }
    return scores

# ============================================================================
# 3. BUILD BASELINE MODELS - FRAUD DATA
# ============================================================================

print("\n" + "="*60)
print("BUILDING BASELINE MODELS - FRAUD DATA")
print("="*60)

# 3.1 Logistic Regression
print("\n3.1 Training Logistic Regression...")
lr_fraud = LogisticRegression(
    max_iter=1000,
    random_state=42,
    class_weight='balanced',
    C=1.0
)
lr_fraud.fit(X_fraud_train, y_fraud_train)

# 3.2 Random Forest
print("\n3.2 Training Random Forest...")
rf_fraud = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf_fraud.fit(X_fraud_train, y_fraud_train)

# 3.3 XGBoost
print("\n3.3 Training XGBoost...")
xgb_fraud = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=10,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False
)
xgb_fraud.fit(X_fraud_train, y_fraud_train)

print("\n Baseline models trained successfully!")

# ============================================================================
# 4. EVALUATE BASELINE MODELS - FRAUD DATA
# ============================================================================

print("\n" + "="*60)
print("EVALUATING MODELS - FRAUD DATA")
print("="*60)

# Store all results
fraud_results = []
fraud_predictions = {}

models_fraud = {
    'Logistic Regression': lr_fraud,
    'Random Forest': rf_fraud,
    'XGBoost': xgb_fraud
}

for name, model in models_fraud.items():
    print(f"\nEvaluating {name}...")
    
    # Evaluate
    metrics, y_pred, y_pred_proba = evaluate_model(
        model, X_fraud_train, y_fraud_train,
        X_fraud_test, y_fraud_test,
        name, 'Fraud Data'
    )
    fraud_results.append(metrics)
    fraud_predictions[name] = {'predictions': y_pred, 'probabilities': y_pred_proba}
    
    # Print metrics
    print(f"  Accuracy: {metrics['Accuracy']:.4f}")
    print(f"  Precision: {metrics['Precision']:.4f}")
    print(f"  Recall: {metrics['Recall']:.4f}")
    print(f"  F1 Score: {metrics['F1 Score']:.4f}")
    print(f"  AUC-ROC: {metrics['AUC-ROC']:.4f}")
    print(f"  AUC-PR: {metrics['AUC-PR']:.4f}")
    print(f"  Confusion Matrix: TP={metrics['True Positives']}, "
          f"FP={metrics['False Positives']}, "
          f"FN={metrics['False Negatives']}, "
          f"TN={metrics['True Negatives']}")

# Create results DataFrame
fraud_results_df = pd.DataFrame(fraud_results)
print("\n" + "="*40)
print("FRAUD DATA - MODEL COMPARISON")
print("="*40)
display(fraud_results_df[['Model', 'Precision', 'Recall', 'F1 Score', 'AUC-ROC', 'AUC-PR']])

# ============================================================================
# 5. CROSS-VALIDATION - FRAUD DATA
# ============================================================================

print("\n" + "="*60)
print("CROSS-VALIDATION - FRAUD DATA")
print("="*60)

fraud_cv_results = []
for name, model in models_fraud.items():
    print(f"\nCross-validating {name}...")
    cv_scores = cross_validate_model(model, X_fraud_train, y_fraud_train, name, 'Fraud Data')
    fraud_cv_results.append(cv_scores)
    print(f"  AUC-ROC: {cv_scores['AUC-ROC_mean']:.4f} (±{cv_scores['AUC-ROC_std']:.4f})")
    print(f"  AUC-PR:  {cv_scores['AUC-PR_mean']:.4f} (±{cv_scores['AUC-PR_std']:.4f})")
    print(f"  F1:      {cv_scores['F1_mean']:.4f} (±{cv_scores['F1_std']:.4f})")

fraud_cv_df = pd.DataFrame(fraud_cv_results)
print("\n" + "="*40)
print("FRAUD DATA - CROSS-VALIDATION RESULTS")
print("="*40)
display(fraud_cv_df[['Model', 'AUC-ROC_mean', 'AUC-ROC_std', 'AUC-PR_mean', 'AUC-PR_std', 'F1_mean', 'F1_std']])

# ============================================================================
# 6. BUILD BASELINE MODELS - CREDIT CARD DATA
# ============================================================================

print("\n" + "="*60)
print("BUILDING BASELINE MODELS - CREDIT CARD DATA")
print("="*60)

# 6.1 Logistic Regression
print("\n6.1 Training Logistic Regression...")
lr_credit = LogisticRegression(
    max_iter=1000,
    random_state=42,
    class_weight='balanced',
    C=1.0
)
lr_credit.fit(X_credit_train, y_credit_train)

# 6.2 Random Forest
print("\n6.2 Training Random Forest...")
rf_credit = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf_credit.fit(X_credit_train, y_credit_train)

# 6.3 XGBoost
print("\n6.3 Training XGBoost...")
xgb_credit = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=10,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False
)
xgb_credit.fit(X_credit_train, y_credit_train)

print("\n Credit card models trained successfully!")

# ============================================================================
# 7. EVALUATE BASELINE MODELS - CREDIT CARD DATA
# ============================================================================

print("\n" + "="*60)
print("EVALUATING MODELS - CREDIT CARD DATA")
print("="*60)

credit_results = []
credit_predictions = {}

models_credit = {
    'Logistic Regression': lr_credit,
    'Random Forest': rf_credit,
    'XGBoost': xgb_credit
}

for name, model in models_credit.items():
    print(f"\nEvaluating {name}...")
    
    # Evaluate
    metrics, y_pred, y_pred_proba = evaluate_model(
        model, X_credit_train, y_credit_train,
        X_credit_test, y_credit_test,
        name, 'Credit Card Data'
    )
    credit_results.append(metrics)
    credit_predictions[name] = {'predictions': y_pred, 'probabilities': y_pred_proba}
    
    # Print metrics
    print(f"  Accuracy: {metrics['Accuracy']:.4f}")
    print(f"  Precision: {metrics['Precision']:.4f}")
    print(f"  Recall: {metrics['Recall']:.4f}")
    print(f"  F1 Score: {metrics['F1 Score']:.4f}")
    print(f"  AUC-ROC: {metrics['AUC-ROC']:.4f}")
    print(f"  AUC-PR: {metrics['AUC-PR']:.4f}")
    print(f"  Confusion Matrix: TP={metrics['True Positives']}, "
          f"FP={metrics['False Positives']}, "
          f"FN={metrics['False Negatives']}, "
          f"TN={metrics['True Negatives']}")

# Create results DataFrame
credit_results_df = pd.DataFrame(credit_results)
print("\n" + "="*40)
print("CREDIT CARD DATA - MODEL COMPARISON")
print("="*40)
display(credit_results_df[['Model', 'Precision', 'Recall', 'F1 Score', 'AUC-ROC', 'AUC-PR']])

# ============================================================================
# 8. CROSS-VALIDATION - CREDIT CARD DATA
# ============================================================================

print("\n" + "="*60)
print("CROSS-VALIDATION - CREDIT CARD DATA")
print("="*60)

credit_cv_results = []
for name, model in models_credit.items():
    print(f"\nCross-validating {name}...")
    cv_scores = cross_validate_model(model, X_credit_train, y_credit_train, name, 'Credit Card Data')
    credit_cv_results.append(cv_scores)
    print(f"  AUC-ROC: {cv_scores['AUC-ROC_mean']:.4f} (±{cv_scores['AUC-ROC_std']:.4f})")
    print(f"  AUC-PR:  {cv_scores['AUC-PR_mean']:.4f} (±{cv_scores['AUC-PR_std']:.4f})")
    print(f"  F1:      {cv_scores['F1_mean']:.4f} (±{cv_scores['F1_std']:.4f})")

credit_cv_df = pd.DataFrame(credit_cv_results)
print("\n" + "="*40)
print("CREDIT CARD DATA - CROSS-VALIDATION RESULTS")
print("="*40)
display(credit_cv_df[['Model', 'AUC-ROC_mean', 'AUC-ROC_std', 'AUC-PR_mean', 'AUC-PR_std', 'F1_mean', 'F1_std']])

# ============================================================================
# 9. VISUALIZATIONS
# ============================================================================

print("\n" + "="*60)
print("CREATING VISUALIZATIONS")
print("="*60)

# 9.1 Confusion Matrices - Fraud Data
print("\n9.1 Generating Confusion Matrices - Fraud Data...")
for name in models_fraud.keys():
    fig = plot_confusion_matrix(
        y_fraud_test, 
        fraud_predictions[name]['predictions'],
        name,
        'Fraud Data'
    )
    plt.savefig(f'../data/processed/confusion_matrix_fraud_{name.replace(" ", "_")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 9.2 Confusion Matrices - Credit Card Data
print("\n9.2 Generating Confusion Matrices - Credit Card Data...")
for name in models_credit.keys():
    fig = plot_confusion_matrix(
        y_credit_test, 
        credit_predictions[name]['predictions'],
        name,
        'Credit Card Data'
    )
    plt.savefig(f'../data/processed/confusion_matrix_credit_{name.replace(" ", "_")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 9.3 ROC and PR Curves - Fraud Data
print("\n9.3 Generating ROC and PR Curves - Fraud Data...")
for name in models_fraud.keys():
    fig = plot_roc_pr_curves(
        y_fraud_test,
        fraud_predictions[name]['probabilities'],
        name,
        'Fraud Data'
    )
    plt.savefig(f'../data/processed/curves_fraud_{name.replace(" ", "_")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# 9.4 ROC and PR Curves - Credit Card Data
print("\n9.4 Generating ROC and PR Curves - Credit Card Data...")
for name in models_credit.keys():
    fig = plot_roc_pr_curves(
        y_credit_test,
        credit_predictions[name]['probabilities'],
        name,
        'Credit Card Data'
    )
    plt.savefig(f'../data/processed/curves_credit_{name.replace(" ", "_")}.png', 
                dpi=300, bbox_inches='tight')
    plt.show()

# ============================================================================
# 10. MODEL COMPARISON AND SELECTION
# ============================================================================

print("\n" + "="*60)
print("MODEL COMPARISON AND SELECTION")
print("="*60)

# Combine results for comparison
fraud_results_df['Dataset'] = 'Fraud Data'
credit_results_df['Dataset'] = 'Credit Card Data'
all_results = pd.concat([fraud_results_df, credit_results_df], ignore_index=True)

print("\nComplete Model Comparison:")
display(all_results[['Dataset', 'Model', 'Precision', 'Recall', 'F1 Score', 'AUC-ROC', 'AUC-PR']])

# 10.1 Best Model Selection
print("\n" + "="*40)
print("BEST MODEL SELECTION")
print("="*40)

# Select best model for Fraud Data
best_fraud_model = fraud_results_df.loc[fraud_results_df['F1 Score'].idxmax()]
print(f"\nBest Model for Fraud Data:")
print(f"  Model: {best_fraud_model['Model']}")
print(f"  F1 Score: {best_fraud_model['F1 Score']:.4f}")
print(f"  AUC-PR: {best_fraud_model['AUC-PR']:.4f}")
print(f"  Recall: {best_fraud_model['Recall']:.4f}")
print(f"  Precision: {best_fraud_model['Precision']:.4f}")

# Select best model for Credit Card Data
best_credit_model = credit_results_df.loc[credit_results_df['F1 Score'].idxmax()]
print(f"\nBest Model for Credit Card Data:")
print(f"  Model: {best_credit_model['Model']}")
print(f"  F1 Score: {best_credit_model['F1 Score']:.4f}")
print(f"  AUC-PR: {best_credit_model['AUC-PR']:.4f}")
print(f"  Recall: {best_credit_model['Recall']:.4f}")
print(f"  Precision: {best_credit_model['Precision']:.4f}")

# ============================================================================
# 11. HYPERPARAMETER TUNING (Optional - XGBoost)
# ============================================================================

print("\n" + "="*60)
print("HYPERPARAMETER TUNING - XGBOOST")
print("="*60)

# Tune XGBoost for Fraud Data
print("\nTuning XGBoost for Fraud Data...")

param_grid = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.05, 0.1, 0.2],
    'n_estimators': [50, 100, 150],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

xgb_tune = XGBClassifier(
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=10
)

# Use a smaller grid for faster tuning
param_grid_small = {
    'max_depth': [4, 6],
    'learning_rate': [0.05, 0.1],
    'n_estimators': [50, 100],
}

print("Performing Grid Search (this may take a few minutes)...")
grid_search = GridSearchCV(
    xgb_tune,
    param_grid_small,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)
grid_search.fit(X_fraud_train, y_fraud_train)

print(f"\nBest parameters: {grid_search.best_params_}")
print(f"Best F1 score: {grid_search.best_score_:.4f}")

# Train final model with best parameters
best_xgb_fraud = XGBClassifier(
    **grid_search.best_params_,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=10
)
best_xgb_fraud.fit(X_fraud_train, y_fraud_train)

# Evaluate tuned model
tuned_metrics, y_pred_tuned, y_pred_proba_tuned = evaluate_model(
    best_xgb_fraud, X_fraud_train, y_fraud_train,
    X_fraud_test, y_fraud_test,
    'XGBoost (Tuned)', 'Fraud Data'
)

print("\nTuned XGBoost Performance:")
print(f"  F1 Score: {tuned_metrics['F1 Score']:.4f}")
print(f"  AUC-PR: {tuned_metrics['AUC-PR']:.4f}")
print(f"  Recall: {tuned_metrics['Recall']:.4f}")
print(f"  Precision: {tuned_metrics['Precision']:.4f}")

# ============================================================================
# 12. SAVE MODELS
# ============================================================================

print("\n" + "="*60)
print("SAVING MODELS")
print("="*60)

# Save best models
joblib.dump(best_xgb_fraud, '../models/best_xgb_fraud.pkl')
joblib.dump(xgb_credit, '../models/best_xgb_credit.pkl')  # or whichever is best

# Save all models
joblib.dump(lr_fraud, '../models/lr_fraud.pkl')
joblib.dump(rf_fraud, '../models/rf_fraud.pkl')
joblib.dump(xgb_fraud, '../models/xgb_fraud.pkl')
joblib.dump(lr_credit, '../models/lr_credit.pkl')
joblib.dump(rf_credit, '../models/rf_credit.pkl')
joblib.dump(xgb_credit, '../models/xgb_credit.pkl')

# Save results
fraud_results_df.to_csv('../data/processed/fraud_model_results.csv', index=False)
credit_results_df.to_csv('../data/processed/credit_model_results.csv', index=False)
all_results.to_csv('../data/processed/all_model_results.csv', index=False)

print("✓ All models saved successfully!")
print("✓ Results saved to CSV files!")
print("\n" + "="*60)
print("TASK 2 COMPLETE!")
print("="*60)

print("\n" + "="*60)
print("FINAL SUMMARY")
print("="*60)

print(f"""
┌─────────────────────────────────────────────────────────────────────┐
│                    MODEL BUILDING SUMMARY                          │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│  FRAUD DATA - Best Model: {best_fraud_model['Model']}                      │
│    F1 Score:  {best_fraud_model['F1 Score']:.4f}                             │
│    AUC-PR:    {best_fraud_model['AUC-PR']:.4f}                             │
│    Recall:    {best_fraud_model['Recall']:.4f}                             │
│    Precision: {best_fraud_model['Precision']:.4f}                             │
│                                                                     │
│  CREDIT CARD DATA - Best Model: {best_credit_model['Model']}                  │
│    F1 Score:  {best_credit_model['F1 Score']:.4f}                             │
│    AUC-PR:    {best_credit_model['AUC-PR']:.4f}                             │
│    Recall:    {best_credit_model['Recall']:.4f}                             │
│    Precision: {best_credit_model['Precision']:.4f}                             │
│                                                                     │
│  MODELS SAVED:                                                     │
│    • best_xgb_fraud.pkl                                            │
│    • best_xgb_credit.pkl                                           │
│    • lr_fraud.pkl, rf_fraud.pkl, xgb_fraud.pkl                    │
│    • lr_credit.pkl, rf_credit.pkl, xgb_credit.pkl                 │
│                                                                     │
│  RESULTS SAVED:                                                    │
│    • fraud_model_results.csv                                       │
│    • credit_model_results.csv                                      │
│    • all_model_results.csv                                         │
│                                                                     │
│  VISUALIZATIONS SAVED:                                             │
│    • Confusion matrices for all models                             │
│    • ROC and PR curves for all models                              │
│                                                                     │
└─────────────────────────────────────────────────────────────────────┘
""")

print("\n Task 2 - Model Building and Training Completed Successfully!")
```
